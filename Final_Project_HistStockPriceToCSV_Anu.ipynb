```python
from yahoo_fin.stock_info import get_data
import yahoo_fin.stock_info as si
import pandas as pd
```

```python
csv_name = 'Banks_hist_stocks.csv'
```

```python
#change date range here, but keep 1day interval
start_date="01/25/2019"
```

```python
end_date="01/25/2022"
```

```python
interval = "1d"
```

```python
df_hist_stock_prices = pd.DataFrame(columns = ['ticker','open','high','low','close','adjclose','volume'])
```

```python

stocks = ['FRBA', 'CSTR', 'MSBI', 'RBB', 'SPFI', 'BFC', 'QCRH', 'BY', 'CBNK', 'CFB','MCBS','OBNK','PFBC','CUBI',
          'CCNE','AMNB','MOFG','FISI','AMAL','WTBA']

for ticker in stocks:
    ticker_df = si.get_data(ticker, start_date=start_date, end_date=end_date, interval=interval)
    df_hist_stock_prices = df_hist_stock_prices.append(ticker_df)
```

```python
df_hist_stock_prices
```

```python
df_hist_stock_prices.index.name = 'Date'
```

```python
df_hist_stock_prices = df_hist_stock_prices.reset_index()
```

```python
df_hist_stock_prices.rename(columns={'ticker': 'Ticker', 'open': 'Open','high':'High','low':'Low','close' : 'Close',\
                                     'adjclose': 'Adj Close','volume': 'Volume'}
                            ,inplace=True)
```

```python
df_hist_stock_prices.reset_index(drop=True, inplace=True)
```

```python
df_hist_stock_prices
```

```python
df_hist_stock_prices.to_csv(csv_name, index=False)
```
